```python
import pandas as pd
import os
```

```python
# Leyendo datos
df = pd.read_excel('data/Dry_Bean_Dataset.xlsx')
df.head()
```

```python
# Estadísticas Significativas
stats = df.describe().transpose()
stats
```

## Conteo de valores nulos

```python
# Conteo de valores nulos por columna
df.isnull().sum()
```

## Dimensiones

```python
# Variable objetivo
y = df["Class"]

# Variables predictoras
X = df.drop(columns=["Class"])

# Verificación de dimensiones
print("Dimensiones de X:", X.shape)
print("Dimensiones de y:", y.shape)
```

## Matriz de correlación entre variables

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = X.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Matriz de correlación")
plt.show()

# Mostrar valores numéricos
print(corr)
```

## Correlación de las variables con la variable objetivo

```python
# Convertir la variable objetivo a valores numéricos para calcular correlación
y_encoded = y.astype("category").cat.codes

# Crear DataFrame temporal para calcular correlaciones
df_corr_target = X.copy()
df_corr_target["Class"] = y_encoded

# Correlación con la variable objetivo
corr_target = df_corr_target.corr()["Class"].drop("Class").sort_values(ascending=False)

# Visualización
plt.figure(figsize=(8,6))
sns.barplot(x=corr_target.values, y=corr_target.index)

plt.title("Correlación de variables con la variable objetivo")
plt.xlabel("Correlación")
plt.ylabel("Variables")

plt.show()

# Mostrar valores numéricos
corr_target
```

## Distribución de la variable objetivo

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x="Class")

plt.title("Distribución de clases")
plt.xlabel("Clase")
plt.ylabel("Frecuencia")

plt.show()

# Mostrar conteo de cada clase
class_counts = df["Class"].value_counts()
print(class_counts)
```

## División de datos en entrenamiento y prueba manteniendo la proporción de clases

Debido al desbalance de clases, se mantendrá la misma proporción del conjunto datos de entrenamiento en el conjunto de datos de prueba.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=7833
)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
# Distribución del conjunto de datos de prueba
plt.figure(figsize=(8,5))
sns.countplot(x=y_test)
plt.title("Distribución de clases en el conjunto de prueba")
plt.xlabel("Clase")
plt.ylabel("Frecuencia")
plt.show()
```

## Guardar conjuntos de entrenamiento y prueba

```python
# Guardando datos divididos
os.makedirs("data/processed", exist_ok=True)

train_df = X_train.copy()
train_df["Class"] = y_train

test_df = X_test.copy()
test_df["Class"] = y_test

train_df.to_csv("data/processed/train.csv", index=False)
test_df.to_csv("data/processed/test.csv", index=False)

print("Archivos guardados en data/processed")
```
