# Solución Primer Parcial

```python
import pandas as pd
```

```python
df = pd.read_csv('study_performance.csv')
df.head()
```

```python
# Columnas con nulos por orden
empty_cols = df.isnull().sum().sort_values(ascending=False)
empty_cols
```

```python
# Matriz de correlación
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_cols].corr()
print(correlation_matrix)
```

```python
# Gráfica de correlación
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Promedios de los scores agrupados por género
df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean()
```

```python
# Promedios de los scores agrupados por grupo étnico
df.groupby('race_ethnicity')[['math_score', 'reading_score', 'writing_score']].mean()
```
