# Imágenes como matrices numéricas

Este cuaderno permite cargar una imagen desde un path local y mostrar que:

- una imagen es un arreglo numérico,
- el `shape` indica alto, ancho y canales,
- cada píxel contiene valores,
- pueden aplicarse transformaciones simples.


```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Escribir aquí el path de la imagen
image_path = "data/usac.jpg"

img = Image.open(image_path)
img_array = np.array(img)

print("Tipo de objeto:", type(img))
print("Shape:", img_array.shape)
print("Tipo de dato:", img_array.dtype)

plt.figure(figsize=(5,5))
plt.imshow(img_array)
plt.axis("off")
plt.title("Imagen original")
plt.show()
```

```python
# Ver una pequeña parte de los datos numéricos
print("Primeros 5x5 píxeles:")
print(img_array[:5, :5])
```

```python
# Separar canales de color si la imagen es RGB
if len(img_array.shape) == 3 and img_array.shape[2] >= 3:
    red = img_array[:, :, 0]
    green = img_array[:, :, 1]
    blue = img_array[:, :, 2]

    print("Shape canal rojo:", red.shape)
    print("Shape canal verde:", green.shape)
    print("Shape canal azul:", blue.shape)

    fig = plt.figure(figsize=(12,4))

    plt.subplot(1,3,1)
    plt.imshow(red, cmap="gray")
    plt.title("Canal rojo")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(green, cmap="gray")
    plt.title("Canal verde")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.imshow(blue, cmap="gray")
    plt.title("Canal azul")
    plt.axis("off")

    plt.tight_layout()
    plt.show()
else:
    print("La imagen no tiene 3 canales RGB.")
```

```python
# Conversión a escala de grises
gray = img.convert("L")
gray_array = np.array(gray)

print("Shape en escala de grises:", gray_array.shape)

plt.figure(figsize=(5,5))
plt.imshow(gray_array, cmap="gray")
plt.axis("off")
plt.title("Escala de grises")
plt.show()
```

```python
# Recorte simple
cropped = img_array[:700, :200]

plt.figure(figsize=(4,4))
plt.imshow(cropped, cmap=None if len(cropped.shape) == 3 else "gray")
plt.axis("off")
plt.title("Recorte")
plt.show()
```

```python
# Inversión de intensidad
inverted = 255 - img_array

plt.figure(figsize=(5,5))
plt.imshow(inverted)
plt.axis("off")
plt.title("Imagen invertida")
plt.show()
```
