# Comparación de métricas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leyendo métricas
lr_metrics = pd.read_csv("metrics/linear_regression_metrics_summary.csv")
dtr_metrics = pd.read_csv("metrics/decission_tree_regression_metrics_summary.csv")
```

```python
lr_metrics
```

```python
dtr_metrics
```

```python
# Concatenando las métricas
metrics = pd.concat([lr_metrics, dtr_metrics], ignore_index=True)
metrics
```

```python
# Gráfica por métrica
def graph_metrics(metrics, metric_type = ""):
    tdf = metrics[["Model","dataset", metric_type]]
    tdf = tdf.pivot(index="dataset", columns="Model", values=metric_type)
    tdf.plot(kind="bar", figsize=(10, 6))
    plt.title(f"Comparación de modelos por {metric_type}")
    plt.ylabel(metric_type)
    plt.xlabel("Dataset")
    plt.xticks(rotation=45)
    plt.legend(title="Model")
    plt.show()
```

## MAE

```python
graph_metrics(metrics,"MAE")
```

## MSE

```python
graph_metrics(metrics,"MSE")
```

## RMSE

```python
graph_metrics(metrics,"RMSE")
```

## R2

```python
graph_metrics(metrics,"R2")
```
